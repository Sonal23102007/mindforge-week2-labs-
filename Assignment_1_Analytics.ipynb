```python
import pandas as pd 
import numpy as np 

np.random.seed(42) 
data = { 
         "Transaction_ID": range(10000, 60000),
         "Store_City": np.random.choice(["Mumbai", "Pune", "Delhi", "Bangalore"], 50000), 
         "Category": np.random.choice(["Electronics", "Clothing", "Groceries"], 50000), 
         "Revenue": np.random.normal(500, 150, 50000)
      } 
df = pd.DataFrame(data)  
df.loc[df.sample(frac=0.05).index, "Revenue"] = np.nan 
df.head()
```

```python
missing_count = df["Revenue"].isna().sum()
print("Missing Revenue values:", missing_count)

median_revenue = df["Revenue"].median()

df["Revenue"] = df["Revenue"].fillna(median_revenue)

print("Missing after imputation:", df["Revenue"].isna().sum())
```

```python
city_revenue = df.groupby("Store_City")["Revenue"].sum()

print(city_revenue)
```
